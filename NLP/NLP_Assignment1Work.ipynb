```python
from nltk.corpus import reuters;
from nltk.corpus import inaugural;
```

```python
#Q2a
for file in reuters.fileids():
    
    words = reuters.words(file)
    # Redefining words list to filter out non-alphabetic characters like space and punctuation
    words = [word for word in words if word.isalpha()]
    char_counter = 0
    #loop to calculate the total number of characters for all eligible words
    for word in words:
        num_chars = len(word)
        char_counter += num_chars
    
    #Calculate total number of sentences and avg word and sentence length for each file
    num_sents = len(reuters.sents(file))
    avg_word_len = int(char_counter/len(words))
    avg_sent_len = int(len(words)/num_sents)
    
    #Print formatted result
    print("File name:",str(file),"||", "Avg word length:",avg_word_len,"||","Avg sentence length:",avg_sent_len)    
```
