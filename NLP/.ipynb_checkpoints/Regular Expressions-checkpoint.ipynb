```python
def stem(word):
    for suffix in ['ing', 'ly', 'ed', 'ious', 'ies', 'ive', 'es', 's', 'ment']:
        if word.endswith(suffix):
            return word[:-len(suffix)]
    return word
```

```python
list1 = ['laptops','processing','loudly']
```

```python
for w in list1:
    print(stem(w))
```
